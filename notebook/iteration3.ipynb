# On test data

```python
textFile = sc.textFile("../data/test_files/book1.txt,../data/test_files/book2.txt")

word_counts = (
    textFile.flatMap(lambda x: x.split(" "))
    .map(lambda x:x.strip('“".?/,;-:'))
    .map(lambda x:x.lower())
    .countByValue()
    )

word_counts = sorted(
    word_counts.items(),
    key=lambda x:x[1],
    reverse=True
    )
```

# On full data

```python
textFile = sc.textFile("../data/books_large_p1.txt,../data/books_large_p2.txt")

word_counts = (
    textFile.flatMap(lambda x: x.split(" "))
    .map(lambda x:x.strip('“".?/,;-:'))
    .map(lambda x:x.lower())
    .countByValue()
    )

word_counts = sorted(
    word_counts.items(),
    key=lambda x:x[1],
    reverse=True
    )
```

```python
for word in word_counts[:100]:
    if word[0] not in stopwords:
        print(word)
```
