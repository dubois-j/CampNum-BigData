```python
from pyspark import SparkContext
from pyspark.sql import SparkSession
import pyspark.sql.functions as sf
import numpy as np
import warnings
import re
```

```python
sc = SparkContext(master='local[1]')
spark = SparkSession.builder.appName('Test').getOrCreate()
```

# On full data

```python
textFile = sc.textFile("hdfs://localhost:9050/data_test/test/books_large_p1.txt,hdfs://localhost:9050/data_test/test/books_large_p2.txt")

word_counts = (
    textFile.flatMap(lambda x: x.split(" "))
    .map(lambda x:x.strip('â€œ".?/,;-:'))
    .map(lambda x:x.lower())
    .countByValue()
    )

word_counts = sorted(
    word_counts.items(),
    key=lambda x:x[1],
    reverse=True
    )
```

```python
rdd = sc.parallelize(word_counts)
rdd.saveAsTextFile("hdfs://localhost:9050/data_test/test/most_words.txt")
```
